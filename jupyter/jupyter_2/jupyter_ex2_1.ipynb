```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('cafe_2year.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
header = next(lines)
x=[]
y=[]
for line in lines:
    if (int(line[0]) % 100) %3 == 0:
        x.append(line[0])
        y.append(line[1])
        
rc('font', family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

plt.bar(x, y)
plt.title('전국의 월별 카페수', fontdict=font1)
plt.xlabel('년')
plt.xlabel('년월')
plt.ylabel('카페수')

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

x = []
y1 = []
y2 = []
y3 = []
y4 = []

for line in lines:
    x.append(line[0])
    y1.append(int(line[1]))
    y2.append(int(line[2]))
    y3.append(int(line[3]))
    y4.append(int(line[4]))

rc('font', family='Malgun Gothic')

font1 = {'size':18, 'color':'green'}

plt.plot(x, y1, label='일반의', color='red', linestyle='-', marker='o')
plt.plot(x, y2, label='인턴', color='blue', linestyle='--', marker='x')
plt.plot(x, y3, label='레지던트', color='yellow', linestyle=':', marker='s')
plt.plot(x, y4, label='전문의', color='purple', linestyle='-.', marker='d')

plt.title('서울 및 광역시 종합병원 의사 수', fontdict=font1)
plt.legend(loc='best')
plt.xlabel('시')
plt.ylabel('의사 수')

plt.show()
        
f.close()
```

```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)
header = next(lines)
seoul = next(lines)
result = int(seoul[1]) + int(seoul[2]) + int(seoul[3]) +int(seoul[4])
doctype = [header[1], header[2], header[3], header[4]]
portion = [int(seoul[1])/result * 100, int(seoul[2])/result * 100, int(seoul[3])/result * 100, int(seoul[4])/result * 100]

rc('font', family='Malgun Gothic')
font1 = {'size':18, 'color':'green'}

plt.title("서울종합병원 의사 유형별 분포", fontdict=font1)
plt.pie(portion, explode = (0, 0.1, 0, 0), labels = doctype, autopct="%.1f%%", shadow=True, startangle = 90)
plt.show()
```
